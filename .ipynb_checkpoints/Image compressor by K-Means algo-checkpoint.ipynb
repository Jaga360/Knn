# Image compression

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
img = cv2.imread("Images/nature.jpg")
# convert colour code 
img = cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
```

```python
plt.imshow(img)
plt.show()
```

```python
h,w,c=img.shape
img=img.reshape(h*w,3)
```

```python
from sklearn.cluster import MiniBatchKMeans
kmeans=MiniBatchKMeans(n_clusters=2)
labels=kmeans.fit_predict(img)
```

```python
centroids=kmeans.cluster_centers_.astype('uint8')
comp_img=centroids[labels]
```

```python
comp_img=comp_img.reshape(h,w,3)
comp_img = cv2.cvtColor(comp_img,cv2.COLOR_BGR2RGB)
plt.imshow(comp_img)
plt.show()
```
